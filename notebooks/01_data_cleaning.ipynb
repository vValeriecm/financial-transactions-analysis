```python
import numpy as np

def format_customer_id(customer_id):
    if pd.isna(customer_id):
        return customer_id
    customer_id = str(customer_id)
    if customer_id == 'C_UNKNOWN':
        return customer_id
    # Check if the ID starts with 'C' and the rest are digits
    if customer_id.startswith('C') and customer_id[1:].isdigit():
        num_part = customer_id[1:]
        # Pad with leading zeros if the numeric part has less than 4 digits
        if len(num_part) < 4:
            return 'C' + num_part.zfill(4)
    return customer_id

df['Customer_ID'] = df['Customer_ID'].apply(format_customer_id)

print("Customer_ID after standardization:")
display(df['Customer_ID'].value_counts(dropna=False).head(10))
```

```python
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
def find_similar_product_names(df, column_name, threshold=70):
    unique_products = df[column_name].dropna().unique()
    similar_pairs = []

    # Convert all products to a list to iterate
    product_list = list(unique_products)

    # To avoid comparing 'A' to 'B' and 'B' to 'A', we iterate with a nested loop
    for i in range(len(product_list)):
        for j in range(i + 1, len(product_list)):
            p1 = product_list[i]
            p2 = product_list[j]
            score = fuzz.ratio(p1.lower(), p2.lower()) # Use lower() for case-insensitive comparison
            if score >= threshold:
                similar_pairs.append((p1, p2, score))
    return similar_pairs

similarities = find_similar_product_names(df, 'Product_Name', threshold=85)

if similarities:
    print(f"Found {len(similarities)} pairs of similar product names (threshold >= 70):")
    for p1, p2, score in similarities:
        print(f"  - '{p1}' and '{p2}' (Similarity: {score})")
else:
    print("No highly similar product names found with the current threshold.")
```

The code above uses `fuzzywuzzy` to find product names that are similar to each other. It iterates through all unique product names and calculates a similarity score between each pair. If the score is above the defined `threshold` (here, 70 out of 100), it prints the pair.
